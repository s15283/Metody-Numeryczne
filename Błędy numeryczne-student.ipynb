***Zadanie 1.***  
Liczbę $\textrm{e}$ możemy zdefiniować jako $\sum\limits_{n=0}^{\infty} \frac{1}{n!}$. Oblicz błąd względny i bezwzględny aproksymacji liczby $\textrm{e}$ poprzez $\textrm{e}^{*}$ w przypadku gdy:  
* $\textrm{e}_1^{*}=\sum\limits_{n=0}^{5} \frac{1}{n!}$    

* $\textrm{e}_2^{*}=\sum\limits_{n=0}^{10} \frac{1}{n!}$  

* $\textrm{e}_3^{*}=\sum\limits_{n=0}^{20} \frac{1}{n!}$  

```python
import math

e = math.e

def approx_e(n):
    s = 0
    for i in range(n + 1):
        s += 1 / math.factorial(i)
    return s

values = [5, 10, 20]

for n in values:
    e_approx = approx_e(n)
    
    error_abs = abs(e - e_approx)
    error_rel = error_abs / e
    
    print(f"n = {n}")
    print(f"e* = {e_approx}")
    print(f"Błąd bezwzględny = {error_abs}")
    print(f"Błąd względny   = {error_rel}")
    print("-" * 40)
```

```python
import math

p = 24
exponents = [0, 1, 2, 5, -1, -3]

print("e | przedział | długość | liczba liczb | ULP | błąd abs | błąd wzgl")
print("-" * 100)

for e in exponents:
    start = 2**e
    end = 2**(e + 1)
    interval_length = end - start
    n_numbers = 2**(p - 1)
    ulp = 2**(e - (p - 1))
    max_abs_error = 0.5 * ulp
    max_rel_error = 2**(-p)
    
    print(f"{e:2} | [{start}, {end}) | {interval_length:8} | {n_numbers:12} | {ulp:.2e} | {max_abs_error:.2e} | {max_rel_error:.2e}")

print("\nWNIOSKI:")

print("- W każdym przedziale liczba dokładnie reprezentowanych liczb jest taka sama: 2^(p-1)")
print("- Długość przedziału wynosi 2^e i rośnie wraz z wykładnikiem")
print("- Odległość między sąsiednimi liczbami (ULP) wynosi 2^(e - (p-1)) i rośnie z e")
print("- Maksymalny błąd bezwzględny wynosi 2^(e - p) i rośnie z e")
print("- Maksymalny błąd względny wynosi 2^(-p) i jest stały dla wszystkich e")
```

* Czy jest to macierz osobliwa? Jak to sprawdzić?

```python
import numpy as np
import math

A = np.array([
    [math.sqrt(2), 1/7],
    [math.pi * math.sqrt(2), math.pi / 7]
])

print("Macierz A:")
print(A)

det_num = np.linalg.det(A)

print("\nWyznacznik numeryczny:")
print(det_num)

det_anal = math.sqrt(2) * (math.pi / 7) - (1/7) * (math.pi * math.sqrt(2))

print("\nWyznacznik analityczny:")
print(det_anal)

print("\nPorównanie:")
print("Różnica:", abs(det_num - det_anal))

print("\nWnioski:")

if abs(det_anal) < 1e-12:
    print("- Macierz jest osobliwa (analitycznie)")
else:
    print("- Macierz nie jest osobliwa (analitycznie)")

if abs(det_num) < 1e-12:
    print("- Numerycznie macierz wygląda na osobliwą")
else:
    print("- Numerycznie macierz może nie być uznana za osobliwą")

print("- Wyniki numeryczne mogą być obarczone błędem zaokrągleń")
print("- Może to prowadzić do błędnej klasyfikacji macierzy")
print("- Szczególnie wrażliwe są macierze źle uwarunkowane lub prawie osobliwe")
```

```python
import math

def f(x):
    return math.cos(x) - x

def f_prim(x):
    return -math.sin(x) - 1

def newton(f, f_prim, x0, eps=1e-10, max_iter=100):
    x_prev = x0
    
    for i in range(max_iter):
        x_n = x_prev - (f(x_prev) / f_prim(x_prev))
        
        print(f"Iteracja {i}: x = {x_n}")
        
        if abs(x_n - x_prev) < eps:
            print("\nZatrzymanie: kolejne przybliżenia są bardzo bliskie.")
            return x_n
        
        if abs(f(x_n)) < eps:
            print("\nZatrzymanie: wartość funkcji bliska zeru.")
            return x_n
        
        x_prev = x_n

    print("\nZatrzymanie: osiągnięto maksymalną liczbę iteracji.")
    return x_n

x0 = 1.0
root = newton(f, f_prim, x0)

print("\nPrzybliżone rozwiązanie:", root)

print("\nWNIOSKI:")
print("- Warunek x_n == x_{n-1} NIE jest bezpieczny w arytmetyce zmiennoprzecinkowej")
print("- Należy stosować warunek z tolerancją (eps)")
print("- Dobre warunki stopu:")
print("  * |x_n - x_{n-1}| < eps")
print("  * |f(x_n)| < eps")
print("- Najlepiej używać obu warunków jednocześnie")
print("- Metoda Newtona dla tej funkcji zbiega do rozwiązania ok. 0.739085")
```
